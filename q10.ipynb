```python
import numpy
import matplotlib.pyplot as plt
import random
```

```python
list1 = [0,0,0,0,0,0]
for i in range (1000):
  val = random.randrange(1,7)
  list1[val-1] = list1[val-1] + 1

print(" occurences of faces of dice in order:",list1)
```

```python
plt.bar([1,2,3,4,5,6],list1)
plt.show()
```
